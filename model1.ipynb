```python
#initializing random number generator with seed

from keras.models import Sequential
from keras.layers import Dense
import numpy

#fix our random seeed reproducibility
numpy.random.seed(7)

```

```python
#loading out pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")
#now splitting into imput and output values
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
#defining our model
#create the model
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
#compiling the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
#fitting the model
model.fit(X, Y, epochs=150, batch_size=10, verbose=2)

#epochs are just the number of iterations through the dataset, 
#while batch size is the number of instances evaluated before a weight update
```

```python
#evaluating our model
scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
#making predictions
predictions = model.predict(X)
#rounded predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
